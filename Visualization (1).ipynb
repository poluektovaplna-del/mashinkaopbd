### Визуализация данных

**Библиотека matplotlib** - это библиотека двумерной графики для языка программирования python, с помощью которой можно создавать высококачественные графики различных форматов. Matplotlib представляет собой модуль-пакет для python.

Подключение библиотек

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
y = x ** 3
```

```python
x
```

```python
y
```

#### Линейный график

```python
figure = plt.figure()
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('Значения x')
axes.set_ylabel('Значения y')
axes.set_title('Линейный график')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y, label='y = x^3')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.title('Линейный график')
plt.legend()
plt.show()
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes2 = figure1.add_axes([0.1, 0.4, 0.5, 0.4])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_xlabel('x array values')
axes1.set_ylabel('y array values')
axes1.set_title('Axes1 title')

axes2.set_xlabel('y array values')
axes2.set_ylabel('x array values')
axes2.set_title('Axes2 title')
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)

axes[0].set_xlabel('x array values')
axes[0].set_ylabel('y array values')
axes[0].set_title('Axes[0] title')


axes[1].plot(y, x)

axes[1].set_xlabel('y array values')
axes[1].set_ylabel('x array values')
axes[1].set_title('Axes[1] title')
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample(10)
```

```python
tips.describe()
```

```python
plt.figure(figsize=(8,6))
plt.hist(tips['total_bill'], bins=10)
plt.xlabel('total_bill')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kde=True, bins=20)
```

```python
sns.jointplot(x='total_bill', y='size', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='husl')
```

```python
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr)
```

```python
sns.heatmap(tips_corr, annot=True)
```

```python
pivot = tips.pivot_table(index=["day"],columns=["size"],values="tip",aggfunc=np.average)
sns.heatmap(pivot)
```

```python
grid = sns.FacetGrid(data=tips, col='time', row='smoker')
grid.map(sns.histplot, 'total_bill')
```

```python
grid = sns.FacetGrid(data=tips, col='time', row='smoker')
grid.map(plt.scatter, 'total_bill', 'tip')
```
